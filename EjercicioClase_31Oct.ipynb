##Ejercicio

```python
#Funcion para graficar:

def grafica(x,y,x1,y1):
  plt.plot(x,y)
  plt.plot(x1,y1)
  plt.show()

#Prueba:
x=np.array([0,1,2,3,4,5,6])
y=np.array([3,2,6,4,8,9,2])

x1=np.array([0,1,2,3,4,5,6])
y1=np.array([10,4,16,12,8,9,0])

grafica(x,y,x1,y1)
```

```python
f1=lambda x:np.cos(x)
f2=lambda x:np.sin(x)
```

```python
x=np.linspace(0,6*np.pi,100)
```

```python
plt.subplot(1,2,1)
plt.plot(x,f1(x),'o:r')
plt.title('Coseno')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,f2(x),'x-k')
plt.title('Seno')
plt.grid()

plt.suptitle('Funciones trigonométricas')

plt.show()
```
