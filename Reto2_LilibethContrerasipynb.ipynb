Crear un programa en Python que pueda convertir unidades de longitud, masa y tiempo entre el Sistema Internacional de Unidades (SI) y el Sistema Inglés.

```python
#MENU:
def opciones():
  print('Menu:')
  print('1- Convertir unidades de longitud')
  print('2- Convertir unidades de masa')
  print('3- Convertir unidades de tiempo')
  print('4- Salir del programa')
```

```python
#Función de menu de conversión de unidades de longitud:
def menuLongitud():
  print('1- Metros -> Centímetros')
  print('2- Centimetros -> Milimetros')
  print('3- Kilometros -> Metros')
  print('4- Kilometros -> centímetros')
  print('5- Yardas -> Metros')
  print('6- Metros -> Pies')
  print('7- Pies -> Centimetros')
  print('8- Kilometros -> Millas')

#Función de menu de conversión de unidades de masa:
def menumasa():
  print('1- Kilogramos -> Gramos')
  print('2- Gramos -> Miligramos')
  print('3- Libras -> Kilogramos')
  print('4- Onzas -> Gramos')

#Función de menu de conversión de unidades de tiempo:
def menutiempo():
  print('1- Horas -> Segundos')
  print('2- Segundos -> Minutos')
  print('3- Minutos -> Horas')
  print('4- Minutos -> Milisegundos')
  print('5- Milisegundos -> Días.')
```

```python
#Funciones de longitud:

#Metros a cm:
def macm(x):
  resultado=x*100
  return resultado

#cm a mm:
def cmamm(x):
  resultado=x*10
  return resultado

#km a m:
def kmam(x):
  resultado=x*1000
  return resultado

#km a cm:
def kmacm(x):
  resultado=x*1000*100
  return resultado

#y a m:
def yam(x):
  resultado=x*0.9144
  return resultado

#m a ft:
def maft(x):
  resultado=x*3.28084
  return resultado

#ft a cm:
def ftacm(x):
  resultado=x*3.28084
  return resultado

#km a mi:
def kmami(x):
  resultado=x*0.621371
  return resultado

```

```python
#Funciones de masa

#kg a g:
def kgag(x):
  resultado=x*1000
  return resultado

#g a mg:
def gamg(x):
  resultado=x*1000
  return resultado

#L a kg:
def lakg(x):
  resultado=x*0.453592
  return resultado

#Onzas a g:
def onzag(x):
  resultado=x*28.3495
  return resultado
```

```python
#Funciones de tiempo:

#Horas a segundos:
def has(x):
  resultado=x*3600
  return resultado

#Segundos a minutos:
def sam(x):
  resultado=x*0.0166667
  return resultado

#MInutos a horas:
def mah(x):
  resultado=x/60
  return resultado

#Minutos a Milisegundos:
def mams(x):
  resultado=x*60000
  return resultado

#Milisegundos a dias
def msad(x):
  resultado=x/8.64e+7
  return resultado
```

```python
#MENU:
print('Bienvenido al convertidor de unidades')

opc=0

while opc != 4:
  print('====================================')
  opciones()
  print('====================================')
  opc=int(input('Operación a realizar: '))
  if opc==1:
    print('\n')
    print('Conversiones de Longitud')
    print('====================================')
    menuLongitud()
    print('====================================')
    lon=int(input('Unidades a convertir: '))
    print('\n')
    if lon==1:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=macm(x)
      print('----------------------------------')
      print(' %f m -> %f cm '%(x,resultado))
      print('----------------------------------')
    elif lon==2:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=cmamm(x)
      print('----------------------------------')
      print('%f cm -> %f mm'%(x,resultado))
      print('----------------------------------')
    elif lon==3:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=kmam(x)
      print('----------------------------------')
      print('%f km -> %f m'%(x,resultado))
      print('----------------------------------')
    elif lon==4:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=kmacm(x)
      print('----------------------------------')
      print('%f km -> %f cm'%(x,resultado))
      print('----------------------------------')
    elif lon==5:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=yam(x)
      print('----------------------------------')
      print('%f yardas -> %f m'%(x,resultado))
      print('----------------------------------')
    elif lon==6:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=maft(x)
      print('----------------------------------')
      print('%f m -> %f ft'%(x,resultado))
      print('----------------------------------')
    elif lon==7:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=ftacm(x)
      print('----------------------------------')
      print('%f ft -> %f cm'%(x,resultado))
      print('----------------------------------')
    elif lon==8:
      x=float(input('Ingresa las unidades a convertir: '))
      resultado=kmami(x)
      print('----------------------------------')
      print('%f km -> %f millas'%(x,resultado))
      print('----------------------------------')
    else:
      print('Opción no valida')
  elif opc == 2:
    menumasa()
    print('====================================')
    mas=int(input('Unidades a convertir: '))
    print('\n')
    if mas==1:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=kgag(x)
      print('----------------------------------')
      print(' %f kg -> %f g '%(x,resultado))
      print('----------------------------------')
    elif mas==2:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=gamg(x)
      print('----------------------------------')
      print('%f g -> %f mg'%(x,resultado))
      print('----------------------------------')
    elif mas==3:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=lakg(x)
      print('----------------------------------')
      print('%f libra -> %f kg'%(x,resultado))
      print('----------------------------------')
    elif mas==4:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=onzag(x)
      print('----------------------------------')
      print('%f onza -> %f g'%(x,resultado))
      print('----------------------------------')
    else:
      print('Opción no valida')
  elif opc==3:
    menutiempo()
    print('====================================')
    time=int(input('Unidades a convertir: '))
    print('\n')
    if time==1:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=has(x)
      print('----------------------------------')
      print(' %f hrs -> %f segundos '%(x,resultado))
      print('----------------------------------')
    elif time==2:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=sam(x)
      print('----------------------------------')
      print('%f segundos -> %f minutos'%(x,resultado))
      print('----------------------------------')
    elif time==3:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=mah(x)
      print('----------------------------------')
      print('%f minutos -> %f hrs'%(x,resultado))
      print('----------------------------------')
    elif time==4:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=mams(x)
      print('----------------------------------')
      print('%f minutos -> %f milisegundos'%(x,resultado))
      print('----------------------------------')
    elif time==5:
      x=float(input('Ingresa las unidades a convertir: '))
      print('\n')
      resultado=msad(x)
      print('----------------------------------')
      print('%f milisegundos -> %f días'%(x,resultado))
      print('----------------------------------')
    else:
      print('Opción no valida')
  elif opc==4:
    print('Saliendo del programa...')
  else:
    print('Intentalo de nuevo :(')


```
